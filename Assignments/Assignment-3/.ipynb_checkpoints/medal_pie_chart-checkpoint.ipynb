```python
#Hector Robinson
#100568706
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#list of labels
label=['USA','CHN','GBR','RUS','KOR','GER','FRA','ITA','HUN','AUS','OTHER']

#list of number of medals
x=[46,38,29,24,13,11,11,8,8,7,107]

#makes the chart circular
plt.axis('equal')

#graph the pie chart, editing the labels, size of chart, and distance of labels
plt.pie(x,labels=label,radius = 3,labeldistance=1.1)

#add title
plt.suptitle("2012 Olympic gold medals 100568706",fontsize=25,y=1.75,color='g')
plt.show()
```
