```python
#Hector Robinson
#100568706


import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline



toronto_file = open('toronto_temperature_2012.dat')
toronto_string = toronto_file.read()
toronto_data= toronto_string.splitlines()

montreal_file = open('montreal_temperature_2012.dat')
montreal_string = montreal_file.read()
montreal_data= montreal_string.splitlines()

vancouver_file = open('vancouver_temperature_2012.dat')
vancouver_string = vancouver_file.read()
vancouver_data= vancouver_string.splitlines()



months = range(1,13)

plt.figure(figsize=(10,8))

plt.plot(months,toronto_data, label = 'Toronto',linewidth=3)
plt.plot(months,montreal_data,label = 'Montreal',linewidth=3)
plt.plot(months,vancouver_data,label = 'Vancouver',linewidth=3)

plt.xlabel('Month', fontsize = 13)
plt.ylabel(' Average monthly temperature [Celcius]',fontsize=13)

plt.suptitle("2012 average monthly temperatures 100568706",fontsize=25,y=1.1,color='g')

plt.legend(loc='lower center', shadow=True, fontsize='x-large')



tick_labels=['Jan','Feb','Mar','Apr','May','June','July','Aug','Sep','Oct','Nov','Dec']


plt.xticks(months,tick_labels)

plt.tick_params(axis='x',length=10)


plt.show()

```
