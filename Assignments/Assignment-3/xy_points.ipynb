```python
#Hector Robinson
#100568706


import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


plt.figure(figsize=(8,8))



#*************************PART A
n = np.loadtxt('xy_points.dat')

x = n[0,0:1000]
y = n[1,0:1000]
#*******************************



#PART B
plt.plot(x,y,'r.')




#*****************************    PART C
xnew = x[np.logical_and(x <-2, np.logical_and(y >= -5, y <= 0))]
ynew = y[np.logical_and(x <-2, np.logical_and(y >= -5, y <= 0))]
        
plt.plot(xnew,ynew,'bo')    
#***********************************



#***********************************PART D
xc=0
yc=5

radius=((x-xc)**2+(y-yc)**2)**(1.0/2.0)

xnewer=x[radius<5]
ynewer=y[radius<5]

plt.plot(xnewer,ynewer,'go')
#****************************************

#Add a title
plt.suptitle("Highlighted points 100568706",fontsize=25,y=1.1,color='y')


plt.show()


        
        
```
