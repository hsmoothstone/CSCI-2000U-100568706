```python
#Hector Robinson

%matplotlib inline
from scipy import misc
import numpy as np
import matplotlib.pyplot as plt

rgb_peppers = misc.face()
misc.imsave('peppers.png', rgb_peppers)


plt.imshow(rgb_peppers)
plt.show()



print("dimensions:  " ,rgb_peppers.shape)

gray_peppers= misc.face(gray=True)
plt.imshow(gray_peppers,cmap=plt.cm.gray)
plt.show()

misc.imsave('peppers-gray.png' , gray_peppers)

#print median and max and min
print("max: ", gray_peppers.max(), "    min: " ,gray_peppers.min())

print("median:   ", gray_peppers[-1/2,-1/2])

#flatten into vector
doubles=gray_peppers.flatten()

    

num_bins = 50
#plot histogram
n, bins, patches = plt.hist(doubles, num_bins, normed=1, facecolor='green', alpha=0.5)
plt.title(r'histogram of grayscale values')

plt.show()
 
#cant figure out how to save histogram as png
#misc.imsave('peppers-histogram.png' , doubles)
```
